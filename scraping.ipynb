```python
import requests
from bs4 import BeautifulSoup

#Página de los mas vendidos
URL = "https://cuspide.com/100-mas-vendidos/"

try:
  response = requests.get(URL, timeout=5) #Pide la página tiempo limite 5 segundos
  response.raise_for_status()  # Esto lanzará una excepción si el código de estado no es 200
except requests.Timeout:
  print("La solicitud excedió el tiempo máximo de espera.")
except requests.RequestException as e:
  print(f"Hubo un error al hacer la solicitud: {e}")

soup = BeautifulSoup(response.text, 'html.parser') #Aplicamos el parser html
libros = []

# Buscamos todas las etiquetas <h3> con la clase específica
for h3_tag in soup.find_all('h3', class_='woocommerce-loop-product__title'):
  # Dentro de cada <h3>, buscamos la etiqueta <a>, tiene esta forma <a href="https://cuspide.com/producto/desarma-este-libro/">DESARMA ESTE LIBRO</a>
  a_tag = h3_tag.find('a')

  # Extraemos el texto (título del libro) y la URL
  titulo = a_tag.text #texto del a
  url = a_tag['href'] #url del a

  #Agregamos el elemento a la lista de diccionarios. (Es una lista, cuyos elementos son diccionarios)
  libros.append({
        'titulo': titulo,
        'url': url
    })
```
